```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10,100)
```

```python
fig,axs = plt.subplots(1,1,figsize=(8,8))
axs.plot(x,sigmoid(x))
axs.set_title('sigmoid')
axs.set(xlabel='X',ylabel='Y')
```
