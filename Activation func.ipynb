```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values for each activation function
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_softmax = softmax(x)

# Plot the activation functions in subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y_sigmoid, color='blue')
axs[0, 0].set_title('Sigmoid')

axs[0, 1].plot(x, y_tanh, color='green')
axs[0, 1].set_title('Tanh')

axs[1, 0].plot(x, y_relu, color='red')
axs[1, 0].set_title('ReLU')

axs[1, 1].plot(x, y_softmax, color='purple')
axs[1, 1].set_title('Softmax')

plt.show()
```
